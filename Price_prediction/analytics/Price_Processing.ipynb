```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.plotly
import plotly
import plotly.graph_objs as go 
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('./data/PricePY.csv')
df = df[::-1]
```

```python
df.head()
```

```python
dollar = df[['Average\nClosing Price','Year Open','Year High','Year Low','Year Close']]
dollar
```

```python
def strsplit(data):
    bushell=data.str.split(pat="$",expand=True)
    MTR = np.array(bushell[1],dtype=float)*36.74*70
    return MTR
```

```python
MTR = strsplit(dollar['Average\nClosing Price'])
MTR_Open = strsplit(dollar['Year Open'])
MTR_Close = strsplit(dollar['Year Close'])
MTR_High = strsplit(dollar['Year High'])
MTR_Low = strsplit(dollar['Year Low'])
Price = pd.DataFrame({'Year':df['Year'],'AVG_Price':MTR,'Year_Open':MTR_Open
                     ,'Year_Close':MTR_Close,'Year_High':MTR_High,'Year_Low':MTR_Low})
Price = Price.reset_index()
Price = Price.drop(['index'],axis=1)
Price.head()

```

```python
def trace(x,y):
    t = go.Scatter(x=np.array(x),y=np.array(y))
    return t
trace0 = trace(Price['Year'],Price['AVG_Price'])
data = [trace0]
plotly.offline.plot(data)
```
