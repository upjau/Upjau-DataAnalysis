```python
import pandas as pd 
import numpy as np
import plotly
import plotly.graph_objs as go
from plotly.offline import plot,init_notebook_mode
init_notebook_mode(connected=True)
```

```python
df_all_data = pd.read_csv('./processed_data/All_data.csv')
df_Price = pd.read_csv('./data/Price_data.csv')
df_all_data
```

```python
def traces(x,y,name,lw,mode):
    trace = go.Scatter(x=np.array(x),y=np.array(y),name=name,line=dict(width=lw),mode=mode)
    return trace
```

```python
trace1 = traces(df_all_data['Market Year'],df_all_data['Production'],name='Production',lw=4,mode='lines+markers')
trace2 = traces(df_all_data['Market Year'],df_all_data['Area Harvested'],name='Area Harevsted',lw=4,mode='lines+markers')
data = [trace1,trace2]
```

```python
plot(data,filename='Area_Production.html')
```

```python
trace1 = traces(df_all_data['Market Year'],df_all_data['Production'],name='Production',lw=4,mode='lines+markers')
trace2 = traces(df_all_data['Market Year'],df_all_data['Domestic Consumption'],name='Consumption',lw=4,mode='lines+markers')
trace3 = traces(df_all_data['Market Year'],df_all_data['Imports'],name='Imports',lw=4,mode='lines+markers')
trace4 = traces(df_all_data['Market Year'],df_all_data['Trade Year Imports from U.S.'],name='U.S Imports',lw=4,mode='lines+markers')
trace5 = traces(df_all_data['Market Year'],df_all_data['Trade Year Exports'],name='Exports',lw=4,mode='lines+markers')
trace6 = traces(df_all_data['Market Year'],df_all_data['AVG_Price'],name='Price',lw=4,mode='lines+markers')

data = [trace1,trace2,trace3,trace4,trace5,trace6]

plot(data,filename='Supply_imports.html')
```

```python
trace1 = traces(df_all_data['Market Year'],df_all_data['AVG_Price'],name='Price',lw=4,mode='lines+markers')
trace2 = traces(df_all_data['Market Year'],df_all_data['Prevyear_close'],name='Previous Year Close',lw=4,mode='lines+markers')

data = [trace1,trace2]
plot(data,filename='Price_Line.html')
```
